```python
data['Girls']=data[data['Sex']=='female']['Name']
data['Girls'].dropna()
l=[]
p=''
for i in data['Girls'].dropna():
    for j in i:
        if j!=',':
            p+=j
        else:
            l.append(p)
            p=''
            break
print(l)
```

```python
max=0
n=''
for t in l:
    if l.count(t)>max:
        max=l.count(t)
        n=t
print(max,n)
```

```python
r=[]
t=''
k=0
for i in data['Girls'].dropna():
    for j in range(len(i)):
        if i[j]=='.':
            k=j
            for n in range(len(i)-k+1):
                if n!=(len(i)-k):
                    t+=i[n+k]
                else:
                    r.append(t)
                    t=''
                    k=0
                    break
                    
print(r)
```
